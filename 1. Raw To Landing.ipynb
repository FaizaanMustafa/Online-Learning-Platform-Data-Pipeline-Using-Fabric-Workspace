### 01. Raw To Landing


```python
today_file ='data'
processed_date = '9999-99-99'
source_account = 'adlsgen7428'
source_container = 'fbproject'
src_relative_path = 'raw'
destination_account = 'adlsgen7428'
```

```python
from pyspark.sql.functions import * 
adls_path = 'abfss://%s@%s.dfs.core.windows.net/%s' % (source_container,source_account,src_relative_path)

print('source storage path is',adls_path)
```

```python
latest_path = f"{adls_path}/{today_file}"
print(latest_path)
```

### Reading the csv File


```python
from pyspark.sql.functions import * 
latest_path = f"{adls_path}/{today_file}"
print(latest_path)
df = spark.read.csv(path = latest_path,header=True,inferSchema=True)

```

```python
from pyspark.sql.functions import * 

if df.count() > 1:
    print('the file has data')
    
    df_new= df.withColumn('Processing_Date',lit(processed_date))

    df_new.write.format('csv').mode('append').option('header','true').partitionBy('Processing_Date').save(f'abfss://fbproject@{destination_account}.dfs.core.windows.net/landing/')
    
    print('Data written to landing zone')

else:
    print('This file contains only header row and no data')
```
